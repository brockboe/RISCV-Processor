### Generate precalculated multiplication results

```python
import numpy as np
f = open('./mult_precomputed.sv', 'w+')
f.write('// autogenerated using precompue_multiplication_results.ipynb\n\n')
f.write('function logic [7:0] m (\n')
f.write('\tinput logic [3:0] address1,\n')
f.write('\tinput logic [3:0] address2\n')
f.write(');\n\n')
f.write('parameter [7:0][7:0] ROM = {\n')

for i in range(15, -1, -1):
    for j in range(15, -1, -1):
        f.write('\t\t8\'d' + str((i * j)) + ', \t // ' + str(i) + ' x ' + str(j) + '\n')
        
f.write('};\n\n')
f.write('return ROM[{address1, address2}];\n')
f.write('endfunction')
f.close()
```

### Generate reduction layer tables

```python
for i in range(15):
    print('rl[' + str(i) + '] =', end='')
    for a in range(8):
        for b in range(8):
            if a + b == i:
                print(' + mult(A[' + str(a) +'], B[' + str(b) +'])', end='')
    print('; ')
```
